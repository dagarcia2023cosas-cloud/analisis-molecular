## Análisis de Propiedades Moleculares

Grafica propiedades calculadas de un conjunto de moléculas.

```python
# Datos de ejemplo: propiedades de moléculas
data = pd.DataFrame({
    "Molécula": ["Metanol", "Etanol", "Propanol", "Butanol"],
    "Peso Molecular": [32.04, 46.07, 60.10, 74.12],
    "Punto de Ebullición (°C)": [64.7, 78.4, 97.2, 117.7],
    "LogP": [-0.74, -0.31, 0.25, 0.88],
})

fig = colab.visualization.plot_molecular_properties(
    data.set_index("Molécula")["Peso Molecular"],
    x_label="Molécula",
    y_label="Peso Molecular (g/mol)",
    title="Peso Molecular de Alcoholes"
)
```

## Interacción con Google Drive

Guarda y carga resultados desde Google Drive.

```python
# Guardar resultados en Drive
data.to_csv("alcoholes.csv", index=False)
colab.drive_utils.save_to_drive("alcoholes.csv", "resultados")
print("Datos guardados en Drive.")
```
